```python
#Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read file. Establish that the file is being read properly with a sample of 5 indexes
purchasing_data = pd.read_json("purchase_data.json")
purchasing_data.head()
```

```python
#Show columns
print(purchasing_data.columns)
print(purchasing_data.count())
```

```python
#Find the amount of players

total_players = pd.DataFrame(purchasing_data)
total_players = (purchasing_data.index)
total_players = pd.DataFrame(total_players)

print ("Total Players: " + str(total_players.count()))
```

```python
#Purchasing Analysis Tool

#Unique Items
unique_amount = purchasing_data['Item ID'].unique()
print ("Amount of Unique Items: " + str(len(pd.value_counts(unique_amount))))

#Total Revenue
total_revenue = (purchasing_data["Price"].sum()) 
print ("Total Revenue: " + str(total_revenue))

#Average Purchase Price
avg_purchase_price = (total_revenue/total_players.count())
avg_purchase_price = (str(avg_purchase_price))
print("Average Purchase Price: " + (avg_purchase_price))

#Total Number of Purchases
total_purchases = purchasing_data['Price'].count()
print ("Total number of purchases: " + str(total_purchases))
```

```python
# Find the percentage and count of gender

#Male % & Count
male_p = (purchasing_data["Gender"] == "Male").sum()
male_percentage = male_p/780
male_count = male_percentage*780
print(male_percentage)
print(male_count)

#Female % & Count
female_p = (purchasing_data["Gender"] == "Female").sum()
female_percentage = female_p/780
female_count = female_percentage*780
print(female_percentage)
print(female_count)

#Non-Disclosed % & Count
non_p = (purchasing_data["Gender"] == "Other / Non-Disclosed").sum()
nd_percentage = non_p/780
nd_count = nd_percentage*780
print(nd_percentage)
print(nd_count)
```

```python
# Purchase Analysis by Gender

# Created a new data frame
gender_purchase = purchasing_data[["Gender", "Item Name","Price"]]
gender_purchase = pd.DataFrame(gender_purchase)
print(gender_purchase.tail(3))
print(" ")

# Average Purchase Price
m_p_c = (purchasing_data["Gender"] == "Male").sum()
male_app = total_revenue/m_p_c

f_p_c = (purchasing_data["Gender"] == "Female").sum()
female_app = total_revenue/f_p_c

n_p_c = (purchasing_data["Gender"] == "Other / Non-Disclosed").sum()
nd_app = total_revenue/n_p_c

print(" ")
print("The Average Purchase Price for Males: ", male_app)
print("The Average Purchase Price for Females: ", female_app)
print("The Average Purchase Price for Non-Disclosed: ", nd_app)
print(" ")

#Purchase Count

print ("Purchase count for Males: ", m_p_c)
print ("Purchase count for Females: ", f_p_c)
print ("Purchase count for Non-Disclosed: ", n_p_c)
print(" ")


#Total Purchase Value

#Nominal Purchase Value
```

```python
#Age Demographics

bins = [0, 10, 20, 30, 40, 50, 60]
group_labels = [" > 10", "10 to 15", "15 to 19", "20 to 29", "30 to 49", "50 < "]

pd.cut(purchasing_data["Age"], bins, labels = group_labels)


purchasing_data["Age Range"] = pd.cut(purchasing_data["Age"], bins, labels = group_labels)
#print(purchasing_data.head())


#Purchase Count

purchase_dem = purchasing_data.groupby("Age Range").count()
purchase_dem = purchasing_data[['Age Range', 'Item Name']]
purchase_dem = purchase_dem.groupby("Age Range").count()
purchase_dem = purchase_dem.rename(columns = {'Item Name': 'Item Count'})
purchase_dem = pd.DataFrame(purchase_dem)

print(purchase_dem)


#Average Purchase Count

purchase_c = purchasing_data.groupby("Age Range")
purchase_c = purchase_c.sum()
purchase_c = pd.DataFrame(purchase_c)
print(purchase_c)

purchase_join = purchase_dem.join(purchase_c)
print(purchase_join)

purchase_join["Average Purchase Count"] = purchase_join["Price"]/purchase_join["Item Count"]
del purchase_join["Item Count"]
del purchase_join["Age"]
del purchase_join["Item ID"]
del purchase_join["Price"]

print(purchase_join)


#Total Purchase Value

tpv = purchasing_data.groupby("Age Range").sum()
del tpv["Age"]
del tpv["Item ID"]
ad_tpv = tpv.rename(columns = {'Price': 'Total Purchase Value'})
print(ad_tpv)

#Normalized Totals

```
